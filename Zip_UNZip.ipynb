# Unzipping Code

```python
import zipfile
import os

def unzip_folder(zip_path, extract_to):
    # Ensure the extraction directory exists
    if not os.path.exists(extract_to):
        os.makedirs(extract_to)
    
    # Open the zip file
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        # Extract all the contents into the specified directory
        zip_ref.extractall(extract_to)
    print(f"Unzipped {zip_path} to {extract_to}")

# Example usage
zip_path = "user-default-efs/CPH CECL Final Runs/CPH Dec Runs/baselines_Dec'24.zip"
extract_to = 'user-default-efs/CPH CECL Final Runs/CPH Dec Runs/'
unzip_folder(zip_path, extract_to)
```

# Zipping Code

```python
import os
import zipfile
    
def zipdir(path, ziph):
    # ziph is zipfile handle
    for root, dirs, files in os.walk(path):
        for file in files:
            ziph.write(os.path.join(root, file), 
                       os.path.relpath(os.path.join(root, file), 
                                       os.path.join(path, '..')))

zip_file = "user-default-efs/CPH CECL Final Runs/Baseline Genration Feb/Baseline_query_Feb'25ip"
file_path = "user-default-efs/CPH CECL Final Runs/Baseline Genration Feb/Baseline_query_Feb'25"
with zipfile.ZipFile(zip_file , 'w', zipfile.ZIP_DEFLATED) as zipf:
    zipdir(file_path, zipf)

```
